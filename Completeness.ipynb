## A forward model for the radius distribution of Earth-like exoplanets
by **Daniel Foreman-Mackey (danfm@nyu.edu; NYU)** & **David W. Hogg (david.hogg@nyu.edu; NYU, MPIA)**

The main idea here is that we are directly modeling Erik's planet candidates as a Poisson sampling with some rate $\Gamma (R_p, P)$. This distribution includes the completeness effects (computed more-or-less identically as in the PNAS paper), geometric transit probability, and parameterized "true" radius and period distributions. We then compute the posterior probability of these parameters conditioned on the candidates in this dataset.

```python
%matplotlib inline
from matplotlib import rcParams
rcParams["savefig.dpi"] = 150
```

```python
import emcee
import triangle
import numpy as np
import pandas as pd
from itertools import izip
import scipy.optimize as op
import matplotlib.pyplot as pl
from scipy.misc import logsumexp
```

These are the numbers read off of Figure 3 in Erik's PNAS paper:

```python
ep_Rp_logbins = 0.5 * np.log10(2) * np.arange(9)
ep_Rp_logbin_cen = 0.5 * (ep_Rp_logbins[1:] + ep_Rp_logbins[:-1])
ep_Rp_values = np.array([12., 14.2, 18.6, 5.9, 1.9, 1.0, 0.9, 0.7])
ep_Rp_pdf = ep_Rp_values / np.sum(ep_Rp_values * (ep_Rp_logbins[1:] - ep_Rp_logbins[:-1]))

ep_p_logbins = np.log10([6.25, 12.5, 25, 50, 100])
ep_p_logbin_cen = 0.5 * (ep_p_logbins[1:] + ep_p_logbins[:-1])
ep_p_values = np.array([8.9, 13.7, 15.8, 15.2])
ep_p_pdf = ep_p_values / np.sum(ep_p_values * (ep_p_logbins[1:] - ep_p_logbins[:-1]))
```

These are the (log10-)period and radius of the candidates in the sample:

```python
candidates = []
for line in open("table_ekoi836.tex"):
    cols = line.split("&")
    candidates.append((np.log10(float(cols[2])), np.log10(float(cols[12]))))
candidates = np.array(candidates, dtype=float)
```

Here we're loading the results of the completeness simulation and reproducing Figure S10 from the supplementary material:

```python
sim = pd.read_hdf("mcDV.h5", "mcDV")
m = sim.found * sim.bDV
logperiod, logRp = np.log10(sim.inj_P), np.log10(sim.inj_Rp)
logperiod_yes, logRp_yes = np.log10(sim.inj_P[m]), np.log10(sim.inj_Rp[m])
logperiod_no, logRp_no = np.log10(sim.inj_P[~m]), np.log10(sim.inj_Rp[~m])
pl.plot(logperiod_yes, logRp_yes, ".b", alpha=0.5, ms=3)
pl.plot(logperiod_no, logRp_no, ".r", alpha=0.5, ms=3);
```

Now, we'll hard-set boundaries for the period/radius box that we'll consider and remove candidates that fall outside of this range:

```python
prng = (6.25, 100)  # (5, 400)
rrng = (1, 16)  # (0.5, 16)
i = (candidates[:, 0] > np.log10(prng[0])) * (candidates[:, 0] < np.log10(prng[1])) * (candidates[:, 1] > np.log10(rrng[0])) * (candidates[:, 1] < np.log10(rrng[1]))
candidates = candidates[i]
```

Here, we're computing the (mean) completeness function in a grid by taking the ratio of recovered injections to total number of injections. This is implemented as a ratio-of-histograms and it would fail if we ever had no injections in a bin so let's not be stupid about that: 

```python
ngrid_p, ngrid_r = 32, 30
logpgrid = np.linspace(np.log10(prng[0]), np.log10(prng[1]), ngrid_p+1)
logrgrid = np.linspace(np.log10(rrng[0]), np.log10(rrng[1]), ngrid_r+1)
img_yes, foo, bar = np.histogram2d(logperiod_yes, logRp_yes, (logpgrid, logrgrid))
img_all, foo, bar = np.histogram2d(logperiod, logRp, (logpgrid, logrgrid))
completeness = img_yes / img_all
lncompleteness = np.log(completeness)
```

Here's our version of Figure 1 from the PNAS paper:

```python
pl.pcolor(logpgrid, logrgrid, completeness.T, cmap="gray") 
pl.plot(candidates[:, 0], candidates[:, 1], ".r", ms=2)
pl.xlim(np.log10(prng))
pl.ylim(np.log10(rrng))
pl.colorbar();
```

Now we'll precompute which bin each candidate falls into. This'll come in handy when we compute the likelihood below.

```python
logpind = np.digitize(candidates[:, 0], logpgrid) - 1
logRpind = np.digitize(candidates[:, 1], logrgrid) - 1
```

Somehow, we'll end up with some candidates that fall in bins of zero completeness. The fuck? The model would always give zero likelihood if we include these points so we'll deal with that by removing them from the sample. This is pure hack that we could talk some more about.

```python
i = np.isfinite(lncompleteness[logpind, logRpind])
logpind = logpind[i]
logRpind = logRpind[i]
candidates = candidates[i]
```

This function is defined here for lack of a better place. It helps plots the results of our analysis later on.

```python
def plot_results(thetas, title=None):
    thetas = np.atleast_2d(thetas)
    fig = pl.figure(figsize=(6, 8))
    ax2 = fig.add_subplot(211)
    ax1 = fig.add_subplot(212)
    for theta in thetas:
        alpha_p, beta_p, logc_p, alpha_Rp, beta_Rp, logc_Rp = theta

        x = logrgrid_cen
        y = ln_broken_powerlaw(x, alpha_Rp, beta_Rp, logc_Rp)
        y -= logsumexp(y + np.log(logrgrid[1:] - logrgrid[:-1]))
        ax1.plot(x, np.exp(y), "k", alpha=0.5)
        
        x = logpgrid_cen
        y = ln_broken_powerlaw(x, alpha_p, beta_p, logc_p)
        y -= logsumexp(y + np.log(logpgrid[1:] - logpgrid[:-1]))
        ax2.plot(x, np.exp(y), "k", alpha=0.5)
    
    ax1.plot(ep_Rp_logbin_cen, ep_Rp_pdf, ".", color="r", ms=5)
    ax1.set_xlim(np.log10(rrng))
    ax1.set_ylim(np.array((-0.1, 1.0))*(ax1.get_ylim()[1]))
    ax1.set_ylabel("$p(\log_{10} R_p)$")
    ax1.set_xlabel("$\log_{10} R_p$")
    ax1.axhline(0.0, color="k", alpha=0.3)
    
    ax2.plot(ep_p_logbin_cen, ep_p_pdf, ".", color="r", ms=5)
    ax2.set_xlim(np.log10(prng))
    ax2.set_ylim(np.array((-0.1, 1.0))*(ax2.get_ylim()[1]))
    ax2.set_ylabel("$p(\log_{10} P)$")
    ax2.set_xlabel("$\log_{10} P$")
    ax2.axhline(0.0, color="k", alpha=0.3)
    
    if title is not None:
        ax2.set_title(title)
```

### The likelihood function

This function gives the integrated transit (ln-)probability as a function of period up to a constant factor:

```python
def transit_lnprob(logp):
    p = 10**logp
    return -2./3 * np.log(p)
```

This function computes an unnormalized broken power law of the form:
$$\Gamma(\log_{10} x; \alpha, \beta, \log_{10}x_\mathrm{c}) = \frac{x^\alpha\,(x_\mathrm{c}+x)^{\beta-\alpha}}{x_\mathrm{c}^\beta}$$

```python
def ln_broken_powerlaw(logx, alpha, beta, logxc):
    x, xc = 10**logx, 10**logxc
    return alpha * np.log(x) + (beta-alpha) * np.log(0.5*(xc+x)) - beta * np.log(xc)
```

Now, we'll pre-compute some factors (the grid cell areas and the bin centers) that will be useful in the likelihood function:

```python
lngridcellarea = np.log(logpgrid[1:]-logpgrid[:-1])[:, None] + np.log(logrgrid[1:]-logrgrid[:-1])[None, :]
logpgrid_cen, logrgrid_cen = 0.5*(logpgrid[1:]+logpgrid[:-1]), 0.5*(logrgrid[1:]+logrgrid[:-1])
```

Then the ln-likelihood function itself. The assumptions (all wrong but maybe not totally insane, natrually) that go into this model are:

1. each candidate is an independent sample of the underlying population of exoplanets,
2. the marginalized completeness function (as a function of period and radius) is known exactly,
3. the radius and period distributions are separable and given by broken power laws in each dimension, and
4. the uncertainties on the measured radii and periods are small compared to variations in the completeness and other distributions.

Under these assumptions, the correctly normalized ln-likelihood is the following:

```python
def lnlike(theta):
    alpha_p, beta_p, logc_p, alpha_Rp, beta_Rp, logc_Rp = theta
    
    # Compute the implied true period distribution.
    lnprob_p = ln_broken_powerlaw(logpgrid_cen, alpha_p, beta_p, logc_p)
        
    # Multiply in the transit probability.
    lnprob_p += transit_lnprob(logpgrid_cen)

    # Compute the implied true radius distribution.
    lnprob_Rp = ln_broken_powerlaw(logrgrid_cen, alpha_Rp, beta_Rp, logc_Rp)

    # Incorporate the completeness function and normalize (this is a
    # probability distribution for P and Rp).
    lnprob = lnprob_p[:, None] + lnprob_Rp[None, :] + lncompleteness
    lnprob -= logsumexp(lnprob + lngridcellarea)

    # Evaluate the ln-likelihood at the candidate positions.
    return np.sum(lnprob[logpind, logRpind])

def nll(theta):
    return -lnlike(theta)
```

We'll apply some broad uniform priors on the parameters of the model:

```python
bounds = [
    (-10, 10),
    (-10, 10),
    np.log10(prng) + np.array([-2, 2]),
    (-10, 10),
    (-10, 10),
    np.log10(rrng) + np.array([-2, 2]),
]
```

Find the maximum likelihood distributions using a bounded optimization. You'll notice that we do hit the boundaries of the allowed region. This has to do with some sort of degeneracy in the problem and when we sample below it doesn't seem to be a problem so let's not sweat it for now...

```python
p0 = [0.0, 0.0, 1.5, 8, -6, 0.1]
results = op.minimize(nll, p0, method="L-BFGS-B", bounds=bounds)
print(results)
```

Using the helper function that we defined above, here are the maximum likelihood distributions compared to Erik's results:

```python
plot_results(results.x, "maximum likelihood");
```

### Posterior sampling

Here we'll define a function that evaluates the ln-prior implied by the set of `bounds` given above and combine that with the likelihood to evaluate the posterior probability up to a constant:

```python
def lnprior(theta):
    if all([b[0] <= t <= b[1] for t, b in izip(theta, bounds)]):
        return 0.0
    return -np.inf

def lnprob(theta):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta)
```

Then, we'll use emcee to sample:

```python
ndim, nwalkers = len(results.x), 32
pos = [p0 + 1e-8 * np.random.randn(ndim) for i in range(nwalkers)]
assert np.all(np.isfinite(map(lnprob, pos)))
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob)
sampler.run_mcmc(pos, 4000);
```

We'll remove 2000 steps as "burn-in" and extract the `2000*32` production samples.

```python
samples = sampler.chain[:, 2000:, :]
samples = samples.reshape((-1, ndim))
```

And here are the posterior constraints on the parameters of the model:

```python
triangle.corner(samples, labels=map("${0}$".format, [r"\alpha_P", r"\beta_P", r"\log_{10}P_c",
                                                     r"\alpha_R", r"\beta_R", r"\log_{10}R_c"]));
```

Finally: here's what the posterior constraints look like:

```python
plot_results(samples[np.random.randint(len(samples), size=24)], "posterior sampling");
```

## A more flexible radius distribution 

Here we'll model the radius distribution as a histogram with 10 bins:

```python
resample = 3
model_bins = logrgrid[::resample]
model_ln_widths = np.log(model_bins[1:] - model_bins[:-1])
model_bin_cen = 0.5*(model_bins[1:]+model_bins[:-1])
model_ln_heights = ln_broken_powerlaw(model_bin_cen, *(results.x[3:]))
model_ln_heights -= logsumexp(model_ln_heights + model_ln_widths)
```

```python
def lnlike2(theta):
    alpha_p, beta_p, logc_p = theta[:3]
    bin_ln_heights = theta[3:]
    norm = logsumexp(bin_ln_heights+model_ln_widths[:-1])
    if norm >= 0.0:
        return -np.inf
    bin_vals = np.append(bin_ln_heights, np.log(1.0-np.exp(norm))-model_ln_widths[-1])
    
    # Compute the implied true period distribution.
    lnprob_p = ln_broken_powerlaw(logpgrid_cen, alpha_p, beta_p, logc_p)
        
    # Multiply in the transit probability.
    lnprob_p += transit_lnprob(logpgrid_cen)

    # Compute the implied true radius distribution.
    lnprob_Rp = bin_vals[np.arange(ngrid_r) // resample]

    # Incorporate the completeness function and normalize (this is a
    # probability distribution for P and Rp).
    lnprob = lnprob_p[:, None] + lnprob_Rp[None, :] + lncompleteness
    lnprob -= logsumexp(lnprob + lngridcellarea)

    # Evaluate the ln-likelihood at the candidate positions.
    return np.sum(lnprob[logpind, logRpind])
```

```python
def lnprior2(theta):
    if all([b[0] <= t <= b[1] for t, b in izip(theta[:3], bounds[:3])]):
        return 0.0
    return -np.inf

def lnprob2(theta):
    lp = lnprior2(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike2(theta)
```

```python
p02 = np.append([0, 0, 1.5], model_ln_heights[:-1])
lnprob2(p02)
print(p02, lnprob2(p02))
```

```python
ndim2, nwalkers2 = len(p02), 32
pos = [p02 + 1e-8 * np.random.randn(ndim2) for i in range(nwalkers2)]
assert np.all(np.isfinite(map(lnprob2, pos)))
sampler2 = emcee.EnsembleSampler(nwalkers2, ndim2, lnprob2)
sampler2.run_mcmc(pos, 4000);
```

```python
pl.plot(sampler2.chain[:, :, 5].T);
```

```python
samples2 = sampler2.chain[:, 2000:, :]
samples2 = samples2.reshape((-1, ndim2))
```

```python
def plot_results2(thetas, title=None):
    thetas = np.atleast_2d(thetas)
    fig = pl.figure(figsize=(6, 8))
    ax2 = fig.add_subplot(211)
    ax1 = fig.add_subplot(212)
    for theta in thetas:
        alpha_p, beta_p, logc_p = theta[:3]
        bin_ln_heights = theta[3:]
        norm = logsumexp(bin_ln_heights+model_ln_widths[:-1])
        bin_vals = np.append(np.exp(bin_ln_heights), (1.0 - np.exp(norm))/np.exp(model_ln_widths[-1]))

        x = logpgrid_cen
        y = ln_broken_powerlaw(x, alpha_p, beta_p, logc_p)
        y -= logsumexp(y + np.log(logpgrid[1:] - logpgrid[:-1]))
        ax2.plot(x, np.exp(y), "k", alpha=0.5)
        
        x = np.array(zip(model_bins[:-1], model_bins[1:])).flatten()
        y = np.array(zip(bin_vals, bin_vals)).flatten()
        ax1.plot(x, y, "k", alpha=0.5)
    
    ax1.plot(ep_Rp_logbin_cen, ep_Rp_pdf, ".", color="r", ms=5)
    ax1.set_xlim(np.log10(rrng))
    ax1.set_ylim(np.array((-0.1, 1.0))*(ax1.get_ylim()[1]))
    ax1.set_ylabel("$p(\log_{10} R_p)$")
    ax1.set_xlabel("$\log_{10} R_p$")
    ax1.axhline(0.0, color="k", alpha=0.3)
    
    ax2.plot(ep_p_logbin_cen, ep_p_pdf, ".", color="r", ms=5)
    ax2.set_xlim(np.log10(prng))
    ax2.set_ylim(np.array((-0.1, 1.0))*(ax2.get_ylim()[1]))
    ax2.set_ylabel("$p(\log_{10} P)$")
    ax2.set_xlabel("$\log_{10} P$")
    ax2.axhline(0.0, color="k", alpha=0.3)
    
    if title is not None:
        ax2.set_title(title)
```

```python
plot_results2(samples2[np.random.randint(len(samples2), size=24)], "posterior sampling");
```
